```python
import pandas as pd
#导入numpy库，并给它一个别名为np
import numpy as np
df1 = pd.read_csv("对应导出.csv", low_memory=False)
#导出的表格需要先筛选出需要的行（装置性材料、发布等），复制到别的表格里，删除没有用的列，转成csv后再用excel打开，转换一下就行，
df2 = pd.read_csv("盛寿.csv") # 读取第二个csv文件
```

```python
df2 = df2.drop(["单价", "交货日期", "交货地点及交换方式"], axis=1) # 删除四列
```

```python
df2 = df2.drop([0,36], axis=0) # 删除行
```

```python
df2['物料编码'] = df2['物料编码'].astype('int64')
```
